# Importing libraries 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Loading Data

```python
data= pd.read_csv("1.01. Simple linear regression.csv")
y=data['SAT'].to_numpy() 
x=data['GPA'].to_numpy()
```

# Ploting data

```python
plt.title("Simple Linear Regression Model")
plt.scatter(x,y)
plt.xlabel("GPA Score")
plt.ylabel("SAT Score")
```

# Training the Model

```python
iteration=200
m=0
c=0
n= len(x)
learning_rate= 0.00212

for i in range(iteration):
    y_pred =m*x+c #Hypothesis 
    d_m = (1/n) * (sum( y_pred -y ) *sum(x)) 
    d_c = (1/n) * (sum(y_pred -y))
    error_squre = (sum((y_pred -y)**2))/ (2*n)
    m= m - learning_rate * d_m
    c= c - learning_rate * d_c
    print("m: {}, c: {}, cost: {}, iteration: {}".format(m,c,error_squre,i))
```

# Ploting the Regression Line

```python
predicted_value=m*x+c
plt.plot(x, predicted_value, "k")
plt.scatter(x,y)
plt.xlabel("GPA Score")
plt.ylabel("Predicted SAT Score")
```

# Prediction 

```python
value = float(input("Enter a value: "))
prediction= m*value+c
print("Prediction is : ", prediction)
```
